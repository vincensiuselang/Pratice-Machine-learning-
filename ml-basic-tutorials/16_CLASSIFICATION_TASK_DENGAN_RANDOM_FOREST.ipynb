## Dataset

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi Features: {X.shape}')
print(f'Class: {set(y)}')
```

## Split data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0
)
```

## Classification dengan RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, # n_estimators=100 berarti model akan membuat 100 pohon keputusan dan menggabungkan prediksinya.
                              random_state=0)

model.fit(X_train, y_train)
```

## Evaluasi model

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
