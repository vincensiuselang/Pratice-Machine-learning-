## Dataset: The MNIST database of handwritten digits

```python
from sklearn.datasets import fetch_openml

X, y = fetch_openml('mnist_784', data_home='./dataset', return_X_y=True)
X.shape

# yang 70.000 itu sample dan yang 784 itu features
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

pos = 1
for data in X[:8]:
    plt.subplot(1, 8, pos)
    plt.imshow(data.shape((28, 28)),
              cmap=cm.Greys_r)
    plt.axis('off')
    pos += 1
 
plt.show()
```

## Split Dataset

```python
# X_train = X[60000]
# y_train = y[60000]
# x_test = X[60000:]
# y_test = X[60000:]

X_train = X[:1000]
y_train = y[:1000]
X_test = X[69000:]
y_test = y[69000:]
```

## Classification dengan SVC (Support Vector Classifier)

```python
from sklearn.svm import SVC

model = SVC(random_state=0)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

## Hyperparameter Tuning dengan GridSeacrhCV

#### Hyperparameter adalah parameter yang kita tentukan sendiri sebelum training.
#### GridSearchCV mencoba semua kombinasi hyperparameter yang kita berikan, lalu memilih kombinasi terbaik berdasarkan akurasi.
#### 

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'kernel': ['rbf', 'poly', 'sigmoid'],
    'C': [0.5, 1, 10, 100],
    'gamma': ['scale', 1, 0.1, 0.01, 0.001]
}

# Ini berarti GridSearchCV akan mencoba 3 pilihan kernel × 4 pilihan C × 5 pilihan gamma = 60 kombinasi hyperparameter.

grid_search = GridSearchCV(estimator=SVC(random_state=0),
                          param_grid=parameters,
                          n_jobs=16, # number of jobs = pada 6 threads prosessor, gw punya 16 jadinay ganti ke 16
                          verbose=1,
                          scoring='accuracy')

grid_search.fit(X_train, y_train)
```

## Hasil tuning dari Hyperparameter

```python
print(f'Best Score: {grid_search.best_score_}')

best_params = grid_search.best_estimator_.get_params()
print(f'Best Parameters:')
for params in parameters:
    print(f'\t{params}: {best_params[params]}')
```

## Predict & Evaluate

```python
y_pred = grid_search.predict(X_test)

print(classification_report(y_test, y_pred))
```
