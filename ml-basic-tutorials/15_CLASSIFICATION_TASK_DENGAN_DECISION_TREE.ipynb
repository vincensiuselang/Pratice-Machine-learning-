## Dataset

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi feature: {X.shape}')
print(f'class: {set(y)}')
```

## Split data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0
)
```

## Classification dengan DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4)
# max_depth pada DecisionTreeClassifier berfungsi untuk membatasi kedalaman maksimal pohon keputusan yang dibuat oleh model.

model.fit(X_train, y_train)
```

## Visualisasi model

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams['figure.dpi']=185 # resolusi gambar
plt.subplots(figsize=(10,10))
tree.plot_tree(model, fontsize=8)
plt.show()
```

## Evaluasi Model

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
